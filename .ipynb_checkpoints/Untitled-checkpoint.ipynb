# hotel screen data

```python
rooms = ['111','101','102','201','202','203','205','301','302','303','305',
'501','502','503','505','601','602','603','605']
lst = []
for room in rooms:
    temp_dct = {'id': room,
    'present': '0',
    'time_reminder': '',
    'overtime': '',
    'name': '',
    'checkin_time': '',
    'checkout_time': '',
    'etc': '',
    'color': '(0,0,0,1)'}
    lst += [temp_dct]
```

```python
rooms = pd.DataFrame(lst)
```

```python
rooms.to_csv('hotel_screen.csv')
```
